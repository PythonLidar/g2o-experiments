```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file0 = 'data/0.jpg'
file1 = 'data/1.jpg'
file2 = 'data/2.jpg'

im0 = cv2.imread(file0)
im1 = cv2.imread(file1)
im2 = cv2.imread(file2)
imgs = [im0, im1, im2]

bgr2rgb = lambda x: cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
fig=plt.figure(figsize=(10, 30))

for i in range(3):
  fig.add_subplot(1,3,i+1)
  plt.imshow(bgr2rgb(imgs[i]))
plt.show()
```

```python
gray = cv2.cvtColor(im0,cv2.COLOR_BGR2GRAY)
corners = cv2.goodFeaturesToTrack(gray, 250,0.01, 70)
corners = np.int0(corners)

print("corners detected: ", len(corners))
imcopy = np.copy(im0)
for i in corners:
  x,y = i.ravel()
  cv2.circle(imcopy,(x,y),50,255,-1)

plt.imshow(imcopy)
plt.show()
```
