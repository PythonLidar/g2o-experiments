### Shi-Tomasi Corner Detection
[link](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_shi_tomasi/py_shi_tomasi.html)

```python
def getCorners(img):
    gray = cv2.cvtColor(im0,cv2.COLOR_BGR2GRAY)
    corners = cv2.goodFeaturesToTrack(gray, 10000,0.01, 10)
    corners = np.int0(corners)
    return corners

corners = getCorners(im0)
print("corners detected: ", len(corners))
imcopy = np.copy(im0)
for i in corners:
  x,y = i.ravel()
  cv2.circle(imcopy,(x,y),20,255,-1)

plt.imshow(imcopy)
plt.show()
```

```python
corners0 = getCorners(imgs[0])
corners1 = getCorners(imgs[1])
corners2 = getCorners(imgs[2])


c_0, des_0 = get_features_orb(imgs[0], corners0)
c_1, des_1 = get_features_orb(imgs[1], corners1)
c_2, des_2 = get_features_orb(imgs[2], corners2)

des_idxs0  = np.array(match_frames(des_0, des_1, c_0, c_1))

des_idxs1  = np.array(match_frames(des_1, des_2, c_1, c_2))


print(np.shape(des_idxs0))
print(des_idxs0[0])

print(np.shape(des_idxs1))
print(des_idxs1[0])
```

```python
focal = 200
K = np.array([ # camera intrinsics
        [focal, 0, im0.shape[1]//2],
        [0, focal, im0.shape[0]//2],
        [0, 0, 1]])

idxs, E = estimate_f_matrix(des_idxs0, c_0, c_1, K)

pts4d, R, t, recovered_coords = get_R_t(E, c_0[des_idxs0[:,0]],c_1[des_idxs0[:,1]], K)
Rt = cvt2Rt(R,t)

idxs1, E1 = estimate_f_matrix(des_idxs1, c_1, c_2, K)

pts4d1, R1, t1, recovered_coords1 = get_R_t(E1, c_1[des_idxs1[:,0]],c_2[des_idxs1[:,1]], K)
Rt1 = cvt2Rt(R1,t1)
```

```python
poses = []
poses.append(np.eye(4))
poses.append(np.dot(poses[-1], Rt))
poses.append(np.dot(poses[-1], Rt1))
print(np.shape(poses))
```

```python
optimizer = g2o.SparseOptimizer()
solver = g2o.BlockSolverSE3(g2o.LinearSolverCholmodSE3())
solver = g2o.OptimizationAlgorithmLevenberg(solver)
optimizer.set_algorithm(solver)

```
